# Gruplama İşlemleri

```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df1 = pd.DataFrame({
    'gruplar' : ['A', 'B', 'C', 'A', 'B', 'C'],
    'veri' : [10, 12, 52, 23, 43, 55]
})
df1
```

```python
df1.groupby('gruplar').count()
```

```python
df1.groupby('gruplar').mean()
```

```python
df1.groupby('gruplar')
```

```python
df2 = sns.load_dataset('planets')
df2.head()
```

```python
df2.groupby('method')
```

```python
df2.groupby('method')['orbital_period'].mean()
```
